```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
plt.figure(figsize=(10, 6))
plt.title('Classification Distribution of Target Variable')
sns.countplot(x='target', data=df)
plt.xlabel('Target Variable')
plt.ylabel('Count')
plt.show()
```

```python
X = df.drop("target", axis=1)
Y = df["target"]
```

Now Apply Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42,stratify=Y)
```

```python
print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)
print("Training Labels   :", Y_train.shape)
print("Testing Labels    :", Y_test.shape)
```

Now apply model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lr_model = LogisticRegression(max_iter=1000)
```

```python
lr_model.fit(X_train, Y_train)
```

```python
y_pred_lr = lr_model.predict(X_test)
```

Now train the model od Deceision Tree

```python
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
dt_model.fit(X_train, Y_train)
```

```python
y_pred_dt = dt_model.predict(X_test)
```

Logistic and Deceision_Tree model successfull

```python
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay
)
```

```python
lr_accuracy = accuracy_score(Y_test, y_pred_lr)
print("Logistic Regression Accuracy:", lr_accuracy)
```

```python
print(classification_report(Y_test, y_pred_lr))
```

```python
ConfusionMatrixDisplay.from_predictions(Y_test, y_pred_lr)

plt.title("Logistic Regression Confusion Matrix")
plt.show()
```

```python
dt_accuracy = accuracy_score(Y_test, y_pred_dt)
print("Decision Tree Accuracy:", dt_accuracy)
```

```python
print(classification_report(Y_test, y_pred_dt))
```

```python
ConfusionMatrixDisplay.from_predictions(Y_test, y_pred_dt)

plt.title("Deceision Tree Confusion Matrix")
plt.show()
```

```python
lr_report = classification_report(Y_test, y_pred_lr, output_dict=True)
```

```python
dt_report= classification_report(Y_test, y_pred_dt, output_dict=True)
```

```python
comp = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1-Score"],
    "Logistic Regression": [
        accuracy_score(Y_test, y_pred_lr),
        lr_report["weighted avg"]["precision"],
        lr_report["weighted avg"]["recall"],
        lr_report["weighted avg"]["f1-score"]
    ],
    "Decision Tree": [
        accuracy_score(Y_test, y_pred_dt),
        dt_report["weighted avg"]["precision"],
        dt_report["weighted avg"]["recall"],
        dt_report["weighted avg"]["f1-score"]
    ]
})

comp
```

```python
comparison_melted = comp.melt(
    id_vars="Metric",
    var_name="Model",
    value_name="Score"
)
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison_melted,
    x="Metric",
    y="Score",
    hue="Model"
)

plt.title("Model Performance Comparison")
plt.xlabel("Evaluation Metrics")
plt.ylabel("Score")
plt.ylim(0, 1.1)
plt.show()
```
